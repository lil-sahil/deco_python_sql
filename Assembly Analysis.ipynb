```python
def get_station_from_comment(comment):
    if comment is None:
        return comment
    else:
        pattern = re.compile(r'(\bop\b|\bstn\b|\bst\b|\bstation\b)(\s|-|#)(\s|-|#)?\s?\d*')
        matches = pattern.finditer(comment)
        for match in  matches:
            
            pattern = re.compile(r'(\d+)')

            matches = pattern.finditer(match[0])

            for match in matches:
                return(match[0])
            
```

```python
df_filtered['Station'] = df_filtered['comment'].apply( get_station_from_comment)
```
